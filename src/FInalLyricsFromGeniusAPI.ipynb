```python
# Change the token for online version 

CLIENT_ACCESS_TOKEN = 'SORRY GET YOUR OWN!'
genius = lg.Genius(CLIENT_ACCESS_TOKEN, 
                   skip_non_songs=True, excluded_terms=["(Remix)", "(Live)"], 
                   remove_section_headers=True, timeout = 120, retries = 10)
#timeout time to resolve the timeout error while getting data from the URL.
```

```python
#Getting the names and the number of songs from users.
#n = str(input("Enter names of artists : "))
#k = int(input("Enter number of songs : "))
    
    #Artists already done : 
    
artists = ['Logic','Eminem','Radiohead','Blink-182','Metallica','Lorde','Pink Floyd','ABBA',
           'Red Hot Chili Peppers','Gorillaz','Tame Impala','Bruce Springsteen','Glass Animals','Iron Maiden',
           'Black Keys','Judas Priest','Coldplay','The Strokes','Oasis','Electric Light Orchestra',
           'Audioslave','Cranberries','Kesha','Linkin Park','The Beatles']    


len(artists)
```

```python
def get_songlyrics(artists, k):  # Write lyrics of k songs by each artist in arr
    count = 0  # Counter
    for name in artists:
        try:
            songs = (genius.search_artist(name, max_songs=k, sort='popularity')).songs
            song_lyrics = [song.lyrics for song in songs]
            file= open(name+".txt","w+")
            file.write(" ".join(song_lyrics))  # Deliminator
            count += 1
            print(f"Songs fetched(from the API):{len(song_lyrics)}")
        except:  #  Broad catch which will give us the name of artist and song that threw the exception
            print(f"We have recieved an exception at {name}: {c}")


get_songlyrics(artists, 100)
```
